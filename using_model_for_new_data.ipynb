# Folosim un model deja antrenat pentru prezicerea unor date noi

```python
# Morariu Tudor

import xgboost
import numpy as np
import pandas as pd

# Incarcam datele 
all_X = pd.read_csv("all_data_fara_power.csv");
```

```python
import pickle

# Incarcam modelul antrenat in notebook-ul "model_training.ipynb"

with open('model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
plt.style.use("dark_background");

# Fac un plot cu datele prezise si avrage-ul pe J, V, A.

ax = plt.figure(figsize=(20, 10)).subplots(2, 1)

lim = 1000;

ax[0].plot(range(lim), model.predict(all_X)[:lim], c='b')
ax[0].legend(["predicted"])

ax[1].plot(range(lim), all_X[["J_avg", "V_avg", "A_avg"]][:lim])
ax[1].legend(("J avrage", "V avrage", "A avrage"), loc='upper left')
```
